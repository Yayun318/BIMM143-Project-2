### Scientific Hypothesis: If the TLR2 receptor of humans is structurally similar to the TLR2 receptor of mice, Mycobacterium Vaccae vaccine might have the same effect of stimulating TLR signaling pathway in humans.

To test the structural similarity between Human and Mouse TLR2 receptors, we first need to align the TLR2 gene sequence of the two species through pariwise sequence. We will also align human and mouse gene sequences with pig and mouse through multiple sequence alignment to rule out random chances that human and mouse share similar sequences in the evolutional progresses. We will get our gene sequences data from NCBI website and download them as fasta file. Then, we will visualize the alignment results through Sequence LOGOS. If there are many shared bases in the sequences, we can assume that Human TLR2 and Mouse TLR2 are structually similar. 

In addition, we will use homology sequence to get the protein structure of Human TLR2 and Mouse TLR2. We will get our protein sequences from NCBI website and download them as fasta file. Then we can use 3D protein measurement to compare different properties of the two species: amino acid sequence, secondary structure (hexlix, turn, and sheet), hydrophobicity. 

If these properties are all similar, we could get to the conclusion that TLR2 receptor of humans is structurally similar to the TLR2 receptor of mice, and thus Mycobacterium Vaccae vaccine might have the same effect of stimulating TLR signaling pathway in huamans.

### Performing Bioinformatics Analysis 1: Multiple Sequence Alignment

Definition:
Multiple sequence alignment (MSA) is a tool used to identify the evolutionary relationships and common patterns between genes. It is generally the alignment of three or more biological sequences (protein or nucleic acid) of similar length. 
In our case, we want to align the TLR2 genes of Human, Mouse, Pig, and Dog, so that in the next step we can use sequence Logos to figure out the similarity between the genes.

```python
# Read in sequence data from the fasta that was saved from the NCBI website.
for filename in glob('*.fna'):
    with open(filename) as f:
        # Create an empty file with the file name added with '-aligned.fna'.
        output = str(filename)
        output += '-aligned.fna'
        # The input for Mafft alignment is the string of the filename.
        in_file = str(filename)
        mafft_cline = MafftCommandline(input=in_file)
        stdout, stderr = mafft_cline()
        # Rewrite the output of the Mafft multiple sequence alignment results to the empty file that we created.
        with open(output, 'w') as handle:
            handle.write(stdout)           
```

### Plotting the Results 1: Sequence Logos

Definition: 
A sequence logo is created from a collection of aligned sequences (from above) and depicts the consensus sequence and diversity of the sequences. It is a great way to visualize large amounts of DNA or protein sequences that share in common a conserved feature. To read the sequence logos, the larger the letter (which represents the bases of DNA sequences) is, the higher frequency the sequences share at the point.

```python
# Read the Mafft multiple sequence alignment results of Human TLR2 gene.
align_Human = AlignIO.read("gene_Human.fna-aligned.fna", "fasta")
# Read the Mafft multiple sequence alignment results of Mouse TLR2 gene.
align_Mouse = AlignIO.read("gene_Mouse.fna-aligned.fna", "fasta")
# Read the Mafft multiple sequence alignment results of Pig TLR2 gene.
align_Pig = AlignIO.read("gene_Pig.fna-aligned.fna", "fasta")
# Read the Mafft multiple sequence alignment results of Dog TLR2 gene.
align_Dog = AlignIO.read("gene_Dog.fna-aligned.fna", "fasta")

# Create a new list called align_list and add each alignment result to the list.
align_list = list()
align_list.append(str(align_Human[0].seq))
align_list.append(str(align_Mouse[0].seq))
align_list.append(str(align_Pig[0].seq))
align_list.append(str(align_Dog[0].seq))
# Print out the align_list to check.
print(align_list)
```

```python
# Create a new list called SequenceLogos_list.
SequenceLogos_list = list()
# Extract and read the first 50 letters in each sequence in the align_list.
for a in align_list:
    first_50 = a[:50]
    SequenceLogos_list.append(first_50)
# Print out the SequenceLogos_list to check.
print(SequenceLogos_list)
```

```python
# Display Sequence Logo.
# Create counts matrix.
Multi_counts_df = lm.alignment_to_matrix(sequences=SequenceLogos_list, to_type='counts', characters_to_ignore='.-X')
# Preview counts dataframe.
Multi_counts_df.head()
# Show full Multi counts.
lm.Logo(Multi_counts_df)
```

```python
# Create a new list called SequenceLogos_list2.
SequenceLogos_list2 = list()
# Extract and read the 1000th to 1050th letters in each sequence in the align_list.
for a in align_list:
    Middle_50 = a[1000:1050]
    SequenceLogos_list2.append(Middle_50)
# Print out the SequenceLogos2_list to check.
print(SequenceLogos_list2)
```

```python
# Display Sequence Logo.
# Create counts matrix.
Multi_counts_df = lm.alignment_to_matrix(sequences=SequenceLogos_list2, to_type='counts', characters_to_ignore='.-X')
# Preview counts dataframe.
Multi_counts_df.head()
# Show full Multi counts.
lm.Logo(Multi_counts_df)
```

### Performing Bioinformatics Analysis 2: Homology Sequence 

Definition:
Homology sequence is the situation where nucleic acid or protein sequences are similar because they have a common evolutionary origin. We want to figure out the functional similarity between Human TLR2 and Mouse TLR2 protein, so if they share common evolutionary origin, it is likely their protein structures are also alike.

```python
# Homology Sequence of Human TLR2. 

# First, create a PDBParser object:
parser = PDBParser()
# Then, create a structure object from a PDB file in the following way (the PDB file in this case is called 1FAT.pdb, PHA-L is a user defined name for the structure):
Human_structure = parser.get_structure("Human_TLR2", "Human.pdb")
# View Biopython structural entities
view = nv.show_biopython(Human_structure)
view
```

### Plotting the Results: 3D Protein Measurement

Protein structure is the three-dimensional arrangement of atoms in an amino acid-chain molecule. With the results from homology sequence, we can visualize the 3D protein structures of Human and Mouse TLR2 and thus could determine the structural differences.

```python
# Polypeptide of Human sequence
polypeptide_builder = CaPPBuilder()
counter = 1
# Print out the polypeptide of Human TLR2 sequence.
for polypeptide in polypeptide_builder.build_peptides(Human_structure):
    Human_seq = polypeptide.get_sequence()
    print(f"Sequence: {counter}, Length: {len(Human_seq)}")
    print(Human_seq)
    counter += 1
```

```python
# Define the polypeptide as analyzed_seq
analyzed_seq = ProteinAnalysis(str(Human_seq))
```

```python
# Analyze Human amino acid sequence.
analyzed_seq.get_amino_acids_percent()
```

```python
# Analyze the secondary structure of human sequence. The three results represent hexlix, turn, and sheet.
analyzed_seq.secondary_structure_fraction()
```

```python
# Analyze the hydrophobicity.
analyzed_seq.gravy()
```

```python
# Polypeptide of Mouse sequence
polypeptide_builder = CaPPBuilder()
counter = 1
# Print out the polypeptide of Mouse TLR2 sequence.
for polypeptide in polypeptide_builder.build_peptides(Mouse_structure):
    Mouse_seq = polypeptide.get_sequence()
    print(f"Sequence: {counter}, Length: {len(Mouse_seq)}")
    print(Mouse_seq)
    counter += 1
```

### Analyzing the Results

In the multiple sequence alignment, we align the sequence among human, mouse, pig, and dog, and then use sequence LOGOS to visualize two segments: first 50 sequences and 1000~1050th seqeunces. Since both the two segments have few shared bases (the same base at one point), the results demonstrate a relatively disparity between Human and Mouse TLR2 gene sequences.

Homology modeling refers to constructing an atomic-resolution model of the "target" protein from its amino acid sequence and an experimental three-dimensional structure of a related homologous protein. By using SWISS Model, we got that the sequence similarity is 0.48 and the coverage is 0.69. Then we use 3D protein measurements to know the specific properties of human and mouse TLR protein. The significant gaps in amino acid sequence percent, structure fraction, and hydrophocbicity between human and mouse TLR protein all demonstrate the different properties between the two proteins. 

Overall, the significant genetic and structural differences between human and mouse TLR2 indicate that Mycobacterium Vaccae vaccine would not have the same effect of stimulating TLR signaling pathway in humans.
